## Univariate Plots
> Summarizes the data and would help find patterns in data.As the name suggests,it would be plot of single variable

```python
import numpy as np # import numpy lib
import pandas as pd # pands lib import
import seaborn as sns # For histograms
```

```python
#defining the source of data.
data_source_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
```

```python
df = pd.read_csv(data_source_url,header = None) # df refers to data frame.IT reads the data as dataframe
df.head() # returns the first entries.No of entries required can be passed as argument.Defaults to 5
```

### Adding Column Names to it.

```python
df.columns = [
    'Type',
    'Alcohol',
    'Malic acid',
    'Ash',
    'Alcalinity of ash',
    'Magnesium',
    'Total phenols',
    'Flavanoids',
    'Nonflavanoid phenols',
    'Proanthocyanins',
    'Color intensity',
    'Hue',
    'OD280/OD315 of diluted wines',
    'Proline'
]
# as found in thier documentation
```

```python
print(df) # Would print the entire data frame with the column names as headers
```

```python
print(df.Alcohol)
```

```python
# Method to plot histograms of any column from a dataframe.
def plotHistogramOf(dataframe,columnname):
    dataframe[columnname].hist() # plots histogram of specified column
```

```python
plotHistogramOf(df,'Alcohol') # this would simply reveal the varaition of specific value  in  a dataframe.
```

### Histogram plots are availabe in pandas to.



```python
df.hist()# this would give the histogram plots for all the columns in this 'df' dataframe.
```

### Density plots.


```python
df.plot(kind='density', subplots=True, layout=(len(df.columns),len(df.columns)), sharex=False) # type of plots,the layout would be number of columns
```

#### Density plots using seaborn

```python
sns.kdeplot(df.Alcohol, shade = True) # Density plot of a single column using seaborn
```
